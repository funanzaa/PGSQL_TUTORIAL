# Report ส่งงาน ehhc

```sql
select SUBSTRING(set_id,1,5) set_id,to_char(updated_at::DATE,'DD-MM-YYYY') as update_at
from assessments 
where updated_at::DATE BETWEEN '2019-10-21' and '2019-11-20' -- AND SUBSTRING(set_id,1,5) = '15196'
order by updated_at asc
```

```sql
with cte as (
    select SUBSTRING(set_id,1,5) set_id,to_char(updated_at::DATE,'DD-MM-YYYY') as update_at
from assessments 
where updated_at::DATE BETWEEN '2019-10-21' and '2019-11-20'
order by updated_at asc
)
SELECT  setup_ehhc.hcode ,setup_ehhc.label ,count(set_id) as count_assessments  --set_id as hcode 
from cte
--inner join master.hospitals m_hos on cte.set_id = m_hos.code
RIGHT join setup_ehhc on cte.set_id = setup_ehhc.hcode
GROUP BY setup_ehhc.label, setup_ehhc.hcode
ORDER BY setup_ehhc.hcode::int asc;
```

```sql
with cte_1 as (
    with cte as (
        select SUBSTRING(set_id,1,5) set_id,to_char(updated_at::DATE,'DD-MM-YYYY') as update_at
    from assessments 
    where updated_at::DATE BETWEEN '2019-10-21' and '2019-11-20'
    order by updated_at asc
    )
    SELECT set_id as hcode ,m_hos.label ,count(set_id) as count_ass
    from cte
    inner join master.hospitals m_hos on cte.set_id = m_hos.code
    GROUP BY set_id,m_hos.label 
    ORDER BY hcode asc
)
SELECT SUM(count_ass) 
FROM cte_1
```
